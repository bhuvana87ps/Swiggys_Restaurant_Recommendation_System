### Objective
The objective of this notebook is to transform the cleaned restaurant dataset
into a numerical format suitable for similarity-based recommendation.

This includes:
- Selecting relevant features
- Encoding categorical features
- Scaling numerical features
- Ensuring index alignment between datasets


```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import OneHotEncoder, StandardScaler
import pickle
```

### Feature Selection

The following features are selected for recommendation modeling:

Categorical Features:
- city
- cuisine

Numerical Features:

```python
categorical_features = ['city', 'cuisine']
numerical_features = ['rating', 'rating_count', 'cost']

df_selected = df[categorical_features + numerical_features]
df_selected.head()
```
